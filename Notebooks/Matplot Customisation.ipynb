```python
%matplotlib inline

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Listing all the styles available for the plot
plt.style.available
```

```python
# Using a style 

plt.style.use("fivethirtyeight")
```

```python
# Making a dumpy data set

x = np.random.randn(10,4)  # Dimension of the np array

x.shape, x.size, x.dtype , type(x)
```

```python
# Exporting it to a pandas dataframe

df = pd.DataFrame(x , columns = ["a" , "b" , "c" , "d"])
df
```

```python
# Plotting directly from the Dataframe

ax = df.plot(kind="bar" , figsize = (10,10))    # made it equal to the ax to use the set property 
ax.set(title="Random number",
       xlabel = "Row number",
       ylabel = "Numbers");

# Incase the legend is not visible 

ax.legend().set_visible(True)
```

```python
# Making earlier plot beautiful 

heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head(5)
```

```python
### Making the Plot all again

# Settings the style of the plot 

plt.style.use("seaborn-whitegrid")

# Creating the canvas 

fig , (ax0,ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10,12),
                               sharex = True)

# Adding the data to the AXES AX0

scatter = ax0.scatter(x = over_50["age"],
            y = over_50["chol"],
            c = over_50["target"],
            cmap = "winter")

ax0.set(title = " Cholestrol Levels vs Age " , 
        ylabel = " Cholestrol Levels ")

ax0.set_ylim(100,600)


ax0.legend(*scatter.legend_elements() , title = "Target")

ax0.axhline(y=over_50["chol"].mean() , 
            linestyle = "--",
            color="r")


# Adding the data to the AXES AX1

ax1.scatter(x = over_50["age"],
            y = over_50["thalach"],
            c = over_50["target"],
            cmap = "winter")

ax1.set(title = " Max Heart Rate vs Age " , 
        xlabel = "Age",
        ylabel = " Cholestrol Levels ");

ax1.set_ylim(60,200)
ax1.set_xlim(50,80)

ax1.legend(*scatter.legend_elements() , title = "Target")

ax1.axhline(y=over_50["thalach"].mean() , 
            linestyle = "--",
           color="r")

# Figure title 

fig.suptitle("Heart Disease Analysis" , fontsize = 16 , fontweight="bold");

```
