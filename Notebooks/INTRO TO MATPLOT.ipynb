#### 0. Importing matplotlib and getting it ready

#### 1. Getting the data ready to be plotted

```python
x = [1,2,3,4,5]
y = [11,66,43,57,86]
```

#### 2. Getting the base canvas ready or creating the figure and assigning the data


```python
# Creating the plot and axes with width and height as 10 each

fig , ax = plt.subplots(figsize = (10,10))

# Adding the data 

ax.plot(x,y)

# Customising the plot 
ax.set(title="Simple plot",
      xlabel = "X-AXIS",
      ylabel= "Y_AXIS")

# Saving the plot as 
fig.savefig("sampleplot.png")
```

### Making subplots a.k.a Multiple Axes

##### OPTION 1 TO MAKE MULTIPLE AXES

```python
# Creating a new canvas

fig , ((ax1,ax2) , (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

# Plotting Data
ax1.plot(x,x/2);
ax2.scatter(np.random.randint(10,size=10),np.random.randint(10,size=10));
ax3.bar(butter_price.keys(),butter_price.values());
ax4.hist(np.random.random(10));
```

##### OPTION 2 TO MAKE MULTIPLE AXES

```python
# Creating the figure / Canvas

fig , ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,10))

# Plotting the data

ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.randint(10,size=10),np.random.randint(10,size=10));
ax[1,0].bar(butter_price.keys(),butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

### Till now we have been using the Stateless method of plotting that is Plotting directly from the data frame

##### Which method to use pyplot vs matplot oo method ?
* When we need quick plots it is ohk to use pyplot
* When we need more advanced plots we use oo method

#### Using the pyplot method for heart disease on heart disease dataframe

```python
# Our Dataframe

heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"]>50]   

# What it does is from heart_disease datafram it goes inside the bracket and check the condition and outputs all 
# member that satisfy it

over_50.head(5)
```

```python
over_50.plot(kind = "scatter",
                x = "age",
                 y = "chol",
                 c = "target");
```

##### Using the pyplot method for heart disease on heart disease dataframe


```python
# OO METHOD MIXED WITH PYPLOT [ The catch is ax = ax makes it less liek pyplot more like oo but still we are lotting directly from over_50 hence is pyplot]

fig , ax = plt.subplots(figsize=(15,6))
over_50.plot(kind="scatter",
            x = "age",
            y = "chol",
            c = "target",
            ax = ax ) # Telling we are using the oo method and we want the plot on axes we created 


# Remember axes is just the space we are working in ( refer the anatomy of a plot for more understanding )
ax.set_xlim(45,100);
```

```python
# OO method from the scratch 

fig , ax = plt.subplots(figsize=(15,6))  # its just a short form of " figure , axes = plt.subplot() "

# Giving another name to the plot 

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Adding labels 

ax.set(title = " Over 50 Heart Disease ",    # Saying set the axis properties to what i say
       xlabel = " Age",
       ylabel = "Cholestrol");

# Adding legend

ax.legend(*scatter.legend_elements() , title="target"); 
# Saying go to scatter and unpack all the elemtns after quering the color paramter in the above plot declartion
# With the title legend 


# Adding horizontal lines to the plot 

ax.axhline(over_50["chol"].mean(),
          linestyle = "--");

```

```python
# How to see the values of the target or any column
over_50.target.values
```

```python
# A subplot of Chol , Thalach and age

fig , (ax0,ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex = True)

## Adding Data to the axis AX0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c = over_50["target"])

## Customising AX0

ax0.set(title="Over 50 Chol",
        ylabel = "Chol")

## Adding legend
ax0.legend(*scatter.legend_elements() , title="Target")

## Adding hline
ax0.axhline(y=over_50["chol"].mean(),
         linestyle = "--");


## ADDING DATA TO AX1

scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

## Customising 

ax1.set(title = "Over 50 Thalac",
        xlabel = "Age",
        ylabel = "Thalach")

## Adding legend

ax1.legend(*scatter.legend_elements() , title = "Title");

## Adding hline

ax1.axhline(y=over_50["thalach"].mean(),
            linestyle = "--");

# Adding title to the figure

fig.suptitle("Heart Disease Analysis" , fontsize = 16 , fontweight="bold");
```
